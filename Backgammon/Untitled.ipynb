```python
import random
def roll_dice():
    roll1, roll2 = random.randint(1, 6), random.randint(1, 6)
    return max(roll1, roll2), min(roll1, roll2)

def generate_dice(turns):
    return [roll_dice() for _ in range(turns)]

deterministic = (0, generate_dice(500))
print(deterministic)
```
